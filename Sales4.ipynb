```python
import pdfplumber
import pandas as pd
import re

def extract_data_from_pdf_revised(pdf_path):
    df = pd.DataFrame(columns=['Doc_Date', 'Time', 'No.', 'All_Bill', 'Product_Code', 'Barcode', 'Quantity', 'Unit_Price'])

    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text = page.extract_text()
            lines = text.split('\n')

            for line in lines:
                if line.strip() == '':
                    continue

                # Revised header pattern match
                match_header = re.match(r'(\d{2}/\d{2}/\d{4}) \((\d{2}:\d{2})\) ทูบคี าซ่า (\S+) ลูกค้าทั\(cid:\d{3}\)วไป (\d+.\d{2})', line)
                if match_header:
                    current_date, current_time, current_no, current_all_bill = match_header.groups()
                    continue

                # Product pattern match
                match_product = re.match(r'(\S+) / (\S+) (.+) (\d+\.\d{2}) (\d+\.\d{2})', line)
                if match_product:
                    product_code, barcode, product_name, quantity, unit_price = match_product.groups()
                    df = df.append({
                        'Doc_Date': current_date,
                        'Time': current_time,
                        'No.': current_no,
                        'All_Bill': current_all_bill,
                        'Product_Code': product_code,
                        'Barcode': barcode,
                        'Quantity': quantity,
                        'Unit_Price': unit_price
                    }, ignore_index=True)

    return df

# Path to the PDF file
pdf_path = 'Sales4.pdf'

# Extracting data from the PDF with the revised function
revised_result_df = extract_data_from_pdf_revised(pdf_path)
print(revised_result_df)
```

```python
import pandas as pd

# สร้าง ExcelWriter
with pd.ExcelWriter('Sales4.xlsx', engine='xlsxwriter') as writer:
    # สร้าง DataFrame ที่มีรูปแบบทศนิยมตามที่ต้องการ
    df_formatted = revised_result_df.copy()
    df_formatted[['All_Bill', 'Quantity', 'Unit_Price']] = df_formatted[['All_Bill', 'Quantity', 'Unit_Price']].round(2)

    # บันทึก DataFrame เป็นไฟล์ Excel
    df_formatted.to_excel(writer, index=False)
```
